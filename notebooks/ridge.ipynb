```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
X = 1.0/(np.arange(1,11) + np.arange(0,10)[:, np.newaxis])
Y = np.ones(10)
```

```python
n_alphas = 250
alphas = np.logspace(-10, -2, n_alphas)

coefic = []
for a in alphas:
    ridge = linear_model.Ridge(alpha=a, fit_intercept=False)
    ridge.fit(X, Y)
    coefic.append(ridge.coef_)
```

```python
ax = plt.gca()
ax.plot(alphas, coefic)
ax.set_xscale("log")
ax.set_xlim(ax.get_xlim()[::-1])
plt.xlabel("alpha")
plt.ylabel("weights")
plt.title('Coeficientes Ridge como uma função regularizadora')
plt.axis('tight')
plt.show()
```
